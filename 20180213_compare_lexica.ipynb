```python
import json, re, unicodedata, unidecode, codecs
import xml.etree.ElementTree as ET
from pprint import pprint
from lxml import etree
from pattern.nl import parse, split, parsetree
from collections import defaultdict, Counter
```

```python
def sort_by_occurence(dictionaryitems):
    sortedlist = []
    for POS in dictionaryitems:
        name=POS[0]
        words = POS[1] #list with tuples. This can be sorted
        sortedwords = sorted(words, key=lambda tup: tup[1],reverse=True)
        newPOS=(name,sortedwords)
        sortedlist.append(newPOS)
    return sortedlist
```

```python
# files
tokentypes = sort_by_occurence(json.load(open('tokentypedict.json')).items())
emos = json.load(open('emodict.json')).items()

#lexica
PA_tree = ET.parse('C:\Users\sternheimam\Dropbox\TNO\Scriptie\Lexica\Pattern.xml')
VU_tree = ET.parse('C:\Users\sternheimam\Dropbox\TNO\Scriptie\Lexica\VU.xml')
```

```python
# convert lexica to sentiment dictionaries
PA_root = PA_tree.getroot()
VU_root = VU_tree.getroot()

# combine lemmas with sentiments in dictionary
lemlist = []
senlist = []
Pattern_lemsent_dict = {}
VU_lemsent_dict = {}

# for pattern
for x in PA_root.findall('./word'):
    Pattern_lemsent_dict[unidecode.unidecode(x.get('form'))] = x.get('polarity')

# for VU
for x in VU_root.findall('./Lexicon/LexicalEntry'):
    for y in x.findall('./Lemma'):
        #print y.get('writtenForm')
        lemlist.append(unidecode.unidecode(y.get('writtenForm')))
    for z in x.findall('./Sense/Sentiment'):
        #print z.get('polarity')
        senlist.append(z.get('polarity'))
for i in range(len(lemlist)):
    VU_lemsent_dict[lemlist[i]] = senlist[i]

    
#-----------------------------------------------------
pprint(Pattern_lemsent_dict)
pprint(VU_lemsent_dict)
```

```python
tuplist = []
for POS in tokentypes:
    if POS[0].startswith('NNP'):
        pass
    else:
        words = POS[1] # I wish to exclude proper nouns, as they are no interesting context words.
        for tup in words:
            tuplist.append(tup)

tokentypesort = sorted(tuplist,key = lambda tup: tup[1], reverse=True)
print tokentypesort
```

...in de VU lijst.
staat het in de VU lijst? geef dan het woord en de score mee naar een DICT
staat het niet in de VU lijst? zet het dan in een andere LIST

...in de PA lijst.
staat het in de PA lijst? geef dan het woord en de score mee naar die DICT
staat het niet in de PA lijst zet het dan in die LIST
stop als LIST 1000 lang is.



```python
novalues=[]
values=defaultdict(list)
count = 0

for wordtype in tokentypesort:
    word =  wordtype[0]
    if wordtype[0] in VU_lemsent_dict:
        VU = VU_lemsent_dict[wordtype[0]]
    else:
        VU = ''
    if wordtype[0] in Pattern_lemsent_dict:
        PA = Pattern_lemsent_dict[wordtype[0]]
    else:
        PA = ''
        
    if VU == '' and PA == '':
        novalues.append(word)
        count+=1
        if len(novalues)>=200:
            break
    elif VU == '' or PA == '':
        values[word].append((VU,PA))
        
with open('novalues.txt',"w") as f:
    f.write('annotate these 200 words: throw out / positive / neutral / negative sentiment\n')
    f.write(" [x/+/0/-] ".join(novalues))
```

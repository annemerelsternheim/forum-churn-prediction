```python
import xml.etree.ElementTree as ET
from pprint import pprint
```

```python
# convert Pattern xml file to sentiment dictionary
tree = ET.parse('C:\Users\sternheimam\Dropbox\TNO\Scriptie\Lexica\Pattern.xml')
root = tree.getroot()

# combine lemmas with sentiments in dictionary
lemlist = []
senlist = []
Pattern_lemsent_dict = {}

for x in root.findall('./word'):
    Pattern_lemsent_dict[x.get('form')] = x.get('polarity')
pprint(Pattern_lemsent_dict)
```

```python
# convert VU xml file to sentiment dictionary
tree = ET.parse('C:\Users\sternheimam\Dropbox\TNO\Scriptie\Lexica\VU.xml')
root = tree.getroot()

# make list of lemmas and list of sentiments
lemlist = []
senlist = []
for x in root.findall('./Lexicon/LexicalEntry'):
    for y in x.findall('./Lemma'):
        #print y.get('writtenForm')
        lemlist.append(y.get('writtenForm'))
    for z in x.findall('./Sense/Sentiment'):
        #print z.get('polarity')
        senlist.append(z.get('polarity'))

# combine lemmas with sentiments in dictionary
VU_lemsent_dict = {}
for i in range(len(lemlist)):
    VU_lemsent_dict[lemlist[i]] = senlist[i]
pprint(VU_lemsent_dict)
```

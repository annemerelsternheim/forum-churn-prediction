# Analysis of THEMATRIX.csv

```python
import pandas as pd
import ast
import numpy as np
import matplotlib.pyplot as plt
```

```python
MATRIX = pd.read_csv(r"C:\Users\sternheimam\Desktop\my-notebook\THEMATRIX.csv",delimiter=";")
```

# AUC analyses

## all means and stds for all cells in matrix

```python
plt.close()
plt.figure(figsize=(20,5))
plt.grid(axis="y")
features=[]
for column in list(MATRIX.columns):
    if column == "Included features":
        for row in MATRIX[column]:
            features.append(row)
    if column!= "Included features":
        col_index = list(MATRIX.columns).index(column)
        means = []
        stds = []
        width = 0.3
        for row in MATRIX[column]:
            N = len(MATRIX[column])
            ind = np.arange(N)
            means.append(np.mean(ast.literal_eval(row)['AUC_scores']))
            stds.append(np.std(ast.literal_eval(row)['AUC_scores']))
        plt.bar(ind+((col_index-2)*width),means,width,yerr=stds)
        for i, v in enumerate(means):
            plt.text(i+((col_index-1.85)*width),v+stds[i]+0.04,"%.3f"%v, rotation=90,ha="center")
plt.ylim(0.5,0.9)
plt.xticks(ind,features,rotation=80)
plt.xlabel("included features")
plt.ylabel("mean AUC")
plt.legend(["churn in 1 month", "churn in 2 months", "churn in 3 months"],loc="upper left")
plt.show()
```

## all means and stds for cells in matrix not resulting from 'static' features

```python
def remove_rows_from_matrix(matrix,feature):
    regex = "[^0-9]"+ str(feature) + "[^0-9]"
    new_matrix = matrix[~matrix['Included features'].str.contains(regex)]
    return new_matrix if len(matrix) == len(new_matrix) else remove_rows_from_matrix(new_matrix,feature)
```

```python
matrix = MATRIX

for x in [4,5,6,7,8]:
    matrix = remove_rows_from_matrix(matrix,x)

plt.close()
plt.figure(figsize=(10,5))
features=[]
for column in list(matrix.columns):
    if column == "Included features":
        for row in matrix[column]:
            features.append(row)
    if column!= "Included features":
        col_index = list(matrix.columns).index(column)
        means = []
        stds = []
        width = 0.25
        for row in matrix[column]:
            N = len(matrix[column])
            ind = np.arange(N)
            means.append(np.mean(ast.literal_eval(row)['AUC_scores']))
            stds.append(np.std(ast.literal_eval(row)['AUC_scores']))
        plt.bar(ind+((col_index-2)*width),means,width,yerr=stds)
        for i, v in enumerate(means):
            plt.text(i+((col_index-2)*width),v+stds[i]+0.01,"%.3f"%v, rotation=90,ha="center", va = "bottom")
plt.ylim(0.5,0.9)
plt.xticks(ind,features,rotation=80)
plt.xlabel("included features")
plt.ylabel("mean AUC")
#    plt.title(title)
plt.legend(["churn in 1 month", "churn in 2 months", "churn in 3 months"],loc="upper left")
plt.show()
```

# making a nice overview of relevant information

```python
matrix = MATRIX
```

```python
new_matrix = pd.DataFrame()
for column in list(matrix.columns):
    churnvalues = []
    learning_rate = []
    n_estimators = []
    max_depth = []
    means = []
    stds = []
    if column == "Included features":
        for row in matrix[column]:
            for n in range(4):
                churnvalues.append(row)    
    if column!= "Included features":
        for row in matrix[column]:
            row = ast.literal_eval(row)
            means.append(np.mean(row['AUC_scores']))
            stds.append(np.std(row['AUC_scores']))
            learning_rate.append(row['parameter_settings']['learning_rate'])
            n_estimators.append(row['parameter_settings']['n_estimators'])
            max_depth.append(row['parameter_settings']['max_depth'])
        for i in range(len(matrix)):
            churnvalues.append(r"%.3f $\pm$ %.3f" %(means[i],stds[i]))
            for v in [learning_rate,n_estimators,max_depth]:
                churnvalues.append(v[i])
    new_matrix[column]=churnvalues
    new_matrix["values"]=[item for sublist in [["AUC","learning\_rate", "n\_estimators", "max\_depth"] for i in range(len(matrix))] for item in sublist]
```

```python
new_matrix.to_csv("shortmatrix.csv", sep="&",index=False)
```
